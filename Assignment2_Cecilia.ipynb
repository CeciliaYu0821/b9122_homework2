### Question 1.1
Extract at least 10 United Nations press release containing the word "crisis". Start with the following seed url: https://press.un.org/en. Notice how press release pages have the "PRESS RELEASE" relative link in the top left corner. Here is an example press release: https://press.un.org/en/2023/sc15431.doc.htm where the “PRESS RELEASE” has the following relative anchor tag:
<a href="/en/press-release" hreflang="en">Press Release</a>
Use this information to determine whether the web page is a press release.

```python
from bs4 import BeautifulSoup
import urllib.request
seed_url = "https://press.un.org/en"
press_crisis = {}
req = urllib.request.Request(seed_url,headers={'User-Agent': 'Mozilla/5.0'})
webpage = urllib.request.urlopen(req).read()
soup = BeautifulSoup(webpage)
press = soup.find_all('div',{"class":"more-link"})[1].find('a')['href'] ## press release page
visit_press = press
i = 0
while visit_press and len(press_crisis) < 10:
    req1 = urllib.request.Request(visit_press,headers={'User-Agent': 'Mozilla/5.0'})
    webpage1 = urllib.request.urlopen(req1).read()
    soup1 = BeautifulSoup(webpage1)
    for tag in soup1.find('div',{"class":"view-content"}).find_all('a', href = True):
        press_url = urllib.parse.urljoin(visit_press,tag['href'])
        req2 = urllib.request.Request(press_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage2 = urllib.request.urlopen(req2).read()
        soup2 = BeautifulSoup(webpage2)
        if soup2.find('a', href='/en/press-release', hreflang='en', string='Press Release') and len(press_crisis) < 10: ## determine press release
            content = soup2.find('div',{"class":"field field--name-body field--type-text-with-summary field--label-hidden field__item"}).get_text()
            print("Visiting: ",press_url)
            if "crisis" in content:
                print(content)
                press_crisis[press_url] = content
                i = i + 1
                with open("1_"+str(i)+".txt","w") as file:
                    file.write(str(soup2))
    next_button = soup1.find('a', title ='Go to next page')
    visit_press = urllib.parse.urljoin(press,next_button['href']) if next_button else None
```

```python
for url in press_crisis.keys():
    print(url)
```

```python
for url,content in press_crisis.items():
    print(url,':')
    print(content)
```

### Question 1.2
Crawl the press room of the European Parliament and extract at least 10 press releases that cover the plenary sessions and contain the word “crisis”. Start with the following seed url: https://www.europarl.europa.eu/news/en/press-room. Notice how press releases related to plenary sessions contain the text “PLENARY SESSIONS” with the following html:
<span class="ep_name">Plenary session</span>.

```python
seed_url = "https://www.europarl.europa.eu/news/en/press-room?contentType=plenary"
press_crisis = {}
visit_press = seed_url
req = urllib.request.Request(visit_press,headers={'User-Agent': 'Mozilla/5.0'})
webpage = urllib.request.urlopen(req).read()
soup = BeautifulSoup(webpage)
i = 0
for tag in soup.find_all('div',{"class":"ep_gridrow-content"})[6].find_all('a', href = True):
    if len(press_crisis) < 10:
        press_url = urllib.parse.urljoin(visit_press,tag['href'])
        req2 = urllib.request.Request(press_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage2 = urllib.request.urlopen(req2).read()
        soup2 = BeautifulSoup(webpage2)
        if soup2.find('span',{"class":"ep_name"} ,string='Plenary session'): ## determine plenary session
            content = soup2.find_all('div',{"class":"ep_gridrow-content"})[2].get_text()
            print("Visiting: ",press_url)
            if "crisis" in content:
                print(content)
                press_crisis[press_url] = content
                i = i + 1
                with open("2_"+str(i)+".txt","w") as file:
                    file.write(str(soup2))
    else:
        break
```

```python
for url in press_crisis.keys():
    print(url)
```

```python
for url,content in press_crisis.items():
    print(url,':')
    print(content)
```
